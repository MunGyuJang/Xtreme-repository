```python
input_ = input()
regex = ['[', ']', ',']

for reg in regex:
    input_ = input_.replace(reg, '')

input_ = list(map(int, input_.split()))

a = set([i for i in range(input_[0], input_[1])])
b = set([i for i in range(input_[2], input_[3])])
c = set([i for i in range(input_[4], input_[5])])

ab, bc, ac = set.intersection(a, b), set.intersection(b, c), set.intersection(a, c)
abc = set.union(ab, bc, ac)

print(len(abc))
```
