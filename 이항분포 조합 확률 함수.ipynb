```python
import math

def percentage(per, n, r):
  f = math.factorial    # 팩토리얼
  result = 0
  for i in range(r, n+1):
    C = f(n) / (f(i) * f(n-i))    # n에서 r이 등장할 수 있는 조합의 수 ex) 3, 2 : OOX OXO XOO = 3
    result += (per**i) * ((1-per)**(n-i)) * C   # n의 r 조합이 나올 확률 분수 곱셈 ex) (4/5의 2제곱 * 1/5의 1제곱 * 조합의 수(C)) + (4/5의 3제곱 * 1/5의 0제곱 * 조합의 수(C)) = 112/125
  
  return result

percentage(0.0035, 300, 1)
```
